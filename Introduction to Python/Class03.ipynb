Indentation in Python:

Indentation refers to the spaces/tabs at the start of a line to define code blocks. Python uses indentation (not {} like other languages) to group statements.

Key Rules:
1. Mandatory: Indentation is required for code blocks (like if, for, functions).
2. Consistency: Use 4 spaces (or 1 tab)per level. Never mix both.
3. Error if missing: Unindented code crashes with IndentationError.

3.) elif Statement

```python
marks = int(input('enter marks'))
if marks >=90:
    print("Grade: A+ (Topper!)")
elif marks >= 80:
    print("Grade: B (Good)")
elif marks >=60:
    print("Grade: C (Average)")
else:
    print("Grade: D (Improve)")
```

Business Case Study: Flipkart Discount Strategy

Problem: Flipkart wants to give discounts based on purchase amount.

• ₹5000+ -> 20% off

• ₹3000+ -> 10% off

• Else -> No Discount

```python
purchase_amount = int(input('Enter the amount: '))

if purchase_amount>= 5000:
    discount = 0.2* purchase_amount
    print(f"Discount: ₹{discount} (20% off)")
elif purchase_amount>=3000:
    discount = 0.1* purchase_amount
    print(f"Discount: ₹{discount} (10% off)")
else:
    print("No Discount Available")

final_price = purchase_amount - discount
print(f"Final Price: ₹{final_price}")
```
