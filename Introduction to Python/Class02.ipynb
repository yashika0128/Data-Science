Create a program that evaluates student results using marks and assigns performance categories like 'Distinction', 'First Class', 'Pass Class', 'Fail'

Bitwise Operators

```python
#These operators work on bits and perform bit-by-bit operations.
```

1.) AND (&)

```python
# Perform bitwise AND.
```

```python
a = 5 #In binary: 0101
b = 3 #In binary: 0011

#Bitwise AND compares each bit position
# 0101 (5)
# 0011 (3)
# ---- 
# 0001 (1) - Only the last bit is 1 in both numbers

result = a & b
print(result) # Output: 1
```

2.) OR (|)

```python
# Performs bitwise OR.
```

```python
#Bitwise OR compares each bit position
# 0101 (5)
# 0011 (3)
# ----
# 0111 (7)

orresult = a|b
print(orresult)
```

```python
# Checks if they do not point to the same object.
```

Case Study: HR Skill Matching System

PS: An HR system needs to check if a candidate has required skills using Identity and Membership Operaors

```python
candidate = {
    'name': 'Ravi',
    'skills': ['Python', 'SQL', 'Communication']
}

# Check if candidate has Python and Communication
if 'Python' in candidate['skills'] and 'Communication' in candidate['skills']:
    print(f"{candidate['name']} is eligible for Data Analyst role")
else:
    print(f"{candidate['name']}needs additional training")
```
